# 16-spliting a string into multiple columns

```python
#import libraries
import pandas as pd
df=pd.DataFrame({'name':['ahmad raza','ali afzal','shahraib'],
                 'location':['lahore, pakistan','sagodha, pakistan','karachi, pakistan']})
df
```

```python
#spliting a column into two coulmn
df.name.str.split(' ',expand=True)
```

```python
#adding those splits into new colums
df[["first name","last name"]]=df.name.str.split(' ',expand=True)
```

```python
df
```

```python
df[["city","country"]]=df.location.str.split(' ',expand=True)
```

```python
df
```

```python
#refine data manipulation
df=df[['first name','last name','city','country']]
df
```
