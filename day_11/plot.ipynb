# plotting in python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data set
kashti=sns.load_dataset("titanic")
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
#kashti.isnall()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived",data=kashti)
```

```python
sns.countplot(x="survived",hue="sex",data=kashti)
```

```python
# scatter plot #matplotlib ki library kas sath
plt.scatter(data=kashti,x="age",y="fare")
```

```python
#uper wala code main kia change kareen ka es ka colour change ho ja 
```

```python
#seaborn ki library ka sath
sns.scatterplot(data=kashti,x="age",y="fare",hue="pclass")
```

```python
sns.barplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
kashti.columns
```

```python
#importanat plot (her column ka row ka sath plot)
sns.pairplot(kashti)
```

```python
#hist plot her numaric value ka plot ban jae ga (bar ko bins bolta haen)
kashti.hist(bins=10,figsize=(9,7),grid=False)
```

```python
sns.boxplot(data=kashti,x="sex",y="fare",hue="who")
```

```python
sns.boxenplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
#how to change x and y title
plt.scatter(data=kashti, x="age", y="fare", c='pclass')
```

```python
#how to change x and y title
```

```python
g=sns.FacetGrid(kashti,hue="survived",col="pclass",margin_titles=True,
                palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter,"fare","age",edgecolor="w").add_legend();
```

```python
g=sns.FacetGrid(kashti,hue="survived",col="sex",margin_titles=True,
                palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter,"fare","age",edgecolor="w").add_legend();
```

```python
g=sns.FacetGrid(kashti,hue="survived",col="who",margin_titles=True,
                palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter,"fare","age",edgecolor="w").add_legend();
```

```python
sns.stripplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
#heat map beautiful plot
corr=kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("heat map of kashti data")
```

```python
sns.violinplot(data=kashti,x="sex",y="fare",hue="pclass")
```
