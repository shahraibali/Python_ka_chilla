EDA is a method used to anaylze and sumarize datasets. in python you can used the pandas library to perform EDA on a dataset.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#find null value(jahan jahan pae null wahan wahan true baki jagan pae false a jae ga)
df.isnull()
#boleen ki shape main ae ga (True,False)
```

```python
df.isnull().sum()
```

```python
#null value in %
df.isnull().sum()*100/len(df)#*100 follow pamdas rule
```

```python
#finding unique value(nunique likhna ka matlb ka total number a jaeen gae unique value ka)
df.nunique()
```

```python
df.columns
```

```python
# visulize the none value in heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#null value ko % ki bajae hum visulae kasa dakh sakta haen
#yticklabel ko false es waja sa kaha gaya hae a agr true karta to rows ki sari value a jati
```

```python
df.drop(columns='deck',inplace=True)
```

```python
df.shape
```

```python
# visulize and explore
sns.countplot(data=df,x='survived')
#count plot main y label ni da sakta
```

```python
sns.barplot(data=df,x='survived',y='fare',hue='who')
#bar plotleya ak side pae numaric value aur dosri side catagorical value
```

```python
sns.countplot(x='pclass',data=df)
```

```python
# data distribution
#bins =starting from zero to a maximum point
sns.histplot(data=df,x='age',bins=30)
#hist plot numaric value pae banta hae unique value ziyada hoon
```

```python
df['age'].hist(bins=30,color='darkred',alpha=0.7)
#alpha graph ki darkness ko show kara ga
```

```python
sns.boxplot(data=df,y='age',x='survived',hue='who')
#box plot banana ka leya ak side pae numaric value aur dosri side catagorical value
```

```python
#x aur y dono numaric value hoon to scatter plot banta hae
sns.scatterplot(x='age',y='fare',data=df,hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(data=df,hue='survived')
```

```python
#stacked barchart/plot
df.groupby(['pclass','who'])['survived'].mean().unstack().plot(kind='bar',stacked=True)
#how to change position of xtickes of bar plot in matplotlib
```

```python
#swarm plot
sns.swarmplot(data=df,x='survived',y='age',hue='who')
```

```python
#point plot
sns.catplot(x='pclass',y='fare',hue='embarked',kind='point',data=df)
```

```python
sns.lmplot(data=df,x='age',y='fare')#assingment ider kia change kareen kA CONDITIONAL ARGUMENT ylim ni lagani)
plt.ylim(0,300)
```
